## Iterate Over a Simple List

```python
list = [1, 2, 3, 4, 5]

for item in list:
    print(item)
```

```python
sales_list = [50, 150, 200, 10]

def purchases_100(sales_list):
    for item in sales_list:
        if item > 100:
            print(item)

purchases_100(sales_list)
```

## Modifying the Function

```python
sales_list = [50, 150, 200, 10, 300, 125]

def purchases_100(sales_list):
    
    count = 0
    for item in sales_list:
        if item > 100:
            count += 1
    return count

purchases_100(sales_list)
```
